```python
import pandas as pd
from pathlib import Path

ROOT = Path.cwd().parents[0]
MODELS_DIR = ROOT / "models"

stat_df = pd.read_csv(MODELS_DIR / "metrics_statistical.csv")
ml_tuned_df = pd.read_csv(MODELS_DIR / "metrics_ml_tuned.csv")

keep_cols = ["Model", "MAE", "RMSE", "sMAPE"]
stat_df = stat_df[keep_cols].drop_duplicates("Model")
ml_tuned_df = ml_tuned_df[keep_cols].drop_duplicates("Model")

tuned = pd.concat([stat_df, ml_tuned_df], ignore_index=True)

for c in ["MAE", "RMSE", "sMAPE"]:
    tuned[c] = pd.to_numeric(tuned[c], errors="coerce")

tuned = tuned.dropna(subset=["MAE", "RMSE", "sMAPE"]).sort_values("MAE").reset_index(drop=True)

display(tuned)

out_path = MODELS_DIR / "metrics_tuned.csv"
tuned.to_csv(out_path, index=False)
print("Saved:", out_path)
```
