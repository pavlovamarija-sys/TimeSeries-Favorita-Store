```python
import pandas as pd
from pathlib import Path

ROOT = Path.cwd().parents[0]          # week_02/ -> project root
MODELS_DIR = ROOT / "models"

stat_df = pd.read_csv(MODELS_DIR / "metrics_statistical.csv")
ml_df   = pd.read_csv(MODELS_DIR / "metrics_ml.csv")

keep_cols = ["Model", "MAE", "RMSE", "sMAPE"]
stat_df = stat_df[keep_cols].drop_duplicates("Model")
ml_df   = ml_df[keep_cols].drop_duplicates("Model")

baseline = pd.concat([stat_df, ml_df], ignore_index=True)

for c in ["MAE", "RMSE", "sMAPE"]:
    baseline[c] = pd.to_numeric(baseline[c], errors="coerce")

baseline = baseline.dropna(subset=["MAE", "RMSE", "sMAPE"]).sort_values("MAE").reset_index(drop=True)

display(baseline)

out_path = MODELS_DIR / "metrics_baseline.csv"
baseline.to_csv(out_path, index=False)
print("Saved:", out_path)
```
